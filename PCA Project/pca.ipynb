```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df=pd.read_csv("/content/houseprice.csv")
df
```

```python
df.isnull().sum()
```

```python
null_val = df.isnull().sum()
null_col = null_val[null_val > 0]
print(null_col)
```

```python
new_df = df.drop(['PoolQC','Fence','MiscFeature','Alley','MasVnrType','FireplaceQu','LotFrontage'], axis=1)
```

```python
null_val = new_df.isnull().sum()
null_col = null_val[null_val > 0]
print(null_col)
```

```python
new_df.info()
```

```python
for i in new_df.select_dtypes(['number']).columns:
  new_df[i] = new_df[i].fillna(new_df[i].median())
```

```python
for i in new_df.select_dtypes(['object']).columns:
   new_df[i]=new_df[i].fillna(new_df[i].mode()[0])
```

```python
new_df.isnull().sum()
```

```python
for i in new_df.select_dtypes(['object']).columns:
  encoder = LabelEncoder()
  encoder.fit(new_df[i].unique())
  new_df[i] = encoder.transform(new_df[i])
```

```python
pca = PCA(n_components=10)
dfs = pca.fit_transform(new_df)
dfs
```

```python
pca_df = pd.DataFrame(data=dfs, columns=[f'PC{i+1}' for i in range(10)])
pca_df['SalePrice'] = new_df['SalePrice']
pca_df
```

```python
from sklearn.ensemble import RandomForestRegressor
x = pca_df.drop('SalePrice', axis=1)
y = pca_df['SalePrice']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model = RandomForestRegressor()
model.fit(x_train,y_train)
prediction = model.predict(x_test)
print(mean_squared_error(prediction,y_test))
print(r2_score(prediction,y_test))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(mean_squared_error(y_pred,y_test))
print(r2_score(y_pred,y_test))
```
